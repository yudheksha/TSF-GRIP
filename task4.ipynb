### Exploratory Data Analysis - Terrorism
#### Perform 'Exploratory Data Analysis' on dataset 'Global Terrorism'

```python
#Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the dataset
d=pd.read_csv("globalterrorismdb_0718dist.csv",low_memory=False)
d.head(10)
```

```python
#Information about dataset
d.info()
```

```python
#Display summary
d.describe()
```

```python
#Retrieve the columns of the dataset
d.columns
```

```python
#Analyze the data
print("Country with most terror attacks:",d['country_txt'].value_counts().idxmax())
print("Region with most terror attacks:",d['region_txt'].value_counts().idxmax())
print("Year with most terror attacks:",d['iyear'].value_counts().idxmax())
print("Group with most terror attacks:",d['gname'].value_counts().index[1])
print("Most Attack Types:",d['attacktype1_txt'].value_counts().idxmax())
```

```python
plt.figure(figsize=(15,8))
x=d['country_txt'].value_counts().index[:10]
y=d['country_txt'].value_counts().values[:10]
sns.barplot(x=x,y=y)
plt.xticks(rotation=90)
plt.show()
```

### Inference: Iraq shows most terror attacks. The above graph shows the top ten countries affected by terrorism.

```python
pd.crosstab(d.iyear,d.region_txt).plot(kind="area",stacked=False,figsize=(15,8))
plt.xlabel("Years")
plt.ylabel("No of terror attacks")
plt.show()
```

### Inference: Region with most terror attacks are the Middle East & North Africa regions, followed by the South Asia region. Australasia & Oceania region has minimum terror attacks.

```python
plt.figure(figsize=(15,8))
plt.title('Terror Attacks across Timeline')
sns.kdeplot(d['iyear'],hue=d['attacktype1_txt'])
plt.xlabel('Years')
plt.ylabel('Frequency of terror attacks')
plt.show()
```

### Inference: Considering the terrorist activities over the years, 2014 has shown to report the maximum cases using bombing/explosion type of attack.

```python
plt.figure(figsize=(15,10))
d['attacktype1_txt'].value_counts().plot.pie(autopct="%1.1f%%")
plt.ylabel("Types of terror attacks")
plt.show()
```

### Inference: Most of the terrorist attacks were in the form of bombing/explosions and armed assaults.

```python
d.plot(kind="scatter",x="nkill",y="nwound",alpha=0.5,figsize=(15,8))
plt.xlabel("No of people killed")
plt.ylabel("No of people wounded")
plt.show()
```

### Inference: Most terrorist attacks have low mortality and injuries, but a few actions have caused many deaths and injuries

```python
countries=d.groupby('country_txt')
US=countries.get_group('United States')
```

```python
US.iyear.plot(kind='hist',bins=30,figsize=(15,8))
plt.xlabel('Years')
plt.ylabel('Frequency')
plt.title('Frequency of US Terror Attacks wrt Years')
plt.show()
```

### Inference: Most terrorist attacks in the US were seen in 1970s and the frequency has more or less decreased in recent years.

```python
plt.figure(figsize=(15,8))
d.groupby('country_txt')['nkill'].sum().sort_values(ascending=False).head(20).plot(kind='bar')
plt.xlabel("Countries")
plt.ylabel("No of people killed")
plt.xticks(rotation=90)
plt.show()
```

### Inference: Maximum no of people killed during terrorist attacks are in Iraq, Afghanistan, Pakistan, Nigeria etc.

```python
plt.figure(figsize=(15,8))
sns.countplot(x=d['targtype1_txt'],order=d['targtype1_txt'].value_counts().index)
plt.xlabel("Target type")
plt.xticks(rotation=90)
plt.show()
```

### Inference: Terror attacks are mostly targeted at private citizens & property, followed by military and police.

```python
plt.figure(figsize=(15,8))
sns.barplot(x=d['gname'].value_counts()[1:11].index,y=d['gname'].value_counts()[1:11].values)
plt.xlabel("Terrorist groups")
plt.xticks(rotation=90)
plt.show()
```

### Inference: The above graph shows the terror groups with the highest attacks.

```python
ind=countries.get_group('India')
```

```python
ind['city'].value_counts().nlargest(10).plot(kind='bar',figsize=(15,8))
plt.show()
```

### Inference: The above graph shows the top ten cities in India with the highest terrorist attacks.

```python
plt.figure(figsize=(15,8))
sns.barplot(x=ind['iyear'],y=ind['nkill'],ci=None)
plt.xlabel("Years")
plt.ylabel("No of people killed in India")
plt.xticks(rotation=90)
plt.show()
```

### Inference: The above graph shows the no of people killed during the terror attacks in India over the years. 1998 shows the maximum mortality.

```python
plt.figure(figsize=(15,8))
sns.barplot(x=ind['gname'].value_counts()[1:11].index,y=ind['gname'].value_counts()[1:11].values)
plt.xlabel("Terrorist groups in India")
plt.xticks(rotation=90)
plt.show()
```

### Inference: The above graph shows the terror groups in India with the highest attacks.
